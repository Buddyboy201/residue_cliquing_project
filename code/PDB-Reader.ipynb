```python
f = open("/Users/kfraga/RESEARCH/ASTRAL-SCOP-WORK/SCOP-2.07/a.1.1.1/d1dlwa_.ent", 'r')
dlwa = f.readlines()
```

```python
f.close()
```

```python
coords = []
for i in dlwa:
    if i[0:4] == 'ATOM':
        new = i.split()
        print(new[6]+' '+new[7]+' '+new[8])
        coords.append([new[6], new[7], new[8]])
```

```python
print(coords)
```

```python
import numpy as np
coords_array = np.array(coords)
```

```python
#print(coords_array)
print(coords_array.shape)
```

```python
import scipy.spatial
```

```python
tri = scipy.spatial.qhull.Delaunay(coords_array)
```

```python
edges = list()
```

```python
import networkx as nx
for n in tri.simplices:
#    print(n)
#    print(n[0])
    edge = sorted([n[0], n[1]])
    edges.append((edge[0], edge[1]))
    edge = sorted([n[0], n[2]])
    edges.append((edge[0], edge[1]))
    edge = sorted([n[0], n[3]])
    edges.append((edge[0], edge[1]))
    edge = sorted([n[1], n[2]])
    edges.append((edge[0], edge[1]))
    edge = sorted([n[1], n[3]])
    edges.append((edge[0], edge[1]))
    edge = sorted([n[2], n[3]])
    edges.append((edge[0], edge[1]))
    
graph = nx.Graph(edges)
```

```python
graph
```

```python
cliques = nx.find_cliques(graph)
```
